```python
# Простая нейронная сеть на Python

weight = 0.1
# weight = 0.5 (пример переобучения)

def neural_network(input, weight):
    pred = input * weight
    return pred
```

```python
# Тестирование нейронной сети

number_of_toes = [8.5, 9.5, 10.0, 9.0]
input = number_of_toes[1]
pred = neural_network(input, weight)
print(f"Предсказание выигрыша: {pred:.2f}%")
```

```python
hours_of_train = [2.5, 5.0, 7.5, 10.0]

pred = neural_network(hours_of_train[2], weight)
print(f"Предсказание сдачи экзаменов: {pred:.2f}%")
```
