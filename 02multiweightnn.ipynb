```python
def w_sum(a, b):
    assert len(a) == len(b)
    # а что делает assert? 
    # Если условие ложно, то программа остановится с ошибкой
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    return output
```

```python
weights = [0.1, 0.2, 0]

def neural_network(input, weights):
    prediction = w_sum(input, weights)
    return prediction
```

```python
toes = [8.5, 9.5, 9.9, 9.0, 3]
wlrec = [0.65, 0.8, 0.4, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

input = [toes[4], wlrec[2], nfans[0]]
pred = neural_network(input, weights)
print(f"Предсказание выигрыша: {pred:.2f}%")
```
