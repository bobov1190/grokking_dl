```python
def elementwise_multiplication(vec_a, vec_b):
    assert (len(vec_a) == len(vec_b))
    return [vec_a[i] * vec_b[i] for i in range(len(vec_a))]
```

```python
def elementwise_addition(vec_a,vec_b):
    assert (len(vec_a) == len(vec_b))
    return [vec_a[i] + vec_b[i] for i in range(len(vec_a))]
```

```python
def vector_sum(vec):
    return sum(vec)
```

```python
def vector_avg(vec):
    return vector_sum(vec) / len(vec)
```

```python
vec_a = [1,2,3]
vec_b = [2,2,4]

print(f"""{elementwise_multiplication(vec_a, vec_b)}
{elementwise_addition(vec_b,vec_a)}
{vector_sum(vec_b)}
{vector_avg(vec_a)}""")
```

```python
def w_sum(inputs, weights):
    return vector_sum(elementwise_multiplication(inputs,weights))
```

```python
toes = [8.5, 9.5, 10.0, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

weights = [0.1, 0.2, 0]

def neural_network(input, weights):
    prediction = w_sum(input, weights)
    return prediction

input = [toes[0], wlrec[0], nfans[0]]
pred = neural_network(input, weights)
print(f"Предсказание выигрыша: {pred:.2f}%")
```
