```python
def w_sum(a, b):
    assert len(a) == len(b)
    # а что делает assert? 
    # Если условие ложно, то программа остановится с ошибкой
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]

    print(output)
```

```python
a = [ 0, 1, 0, 1]
b = [ 1, 0, 1, 0]
c = [ 0, 1, 1, 0]
d = [.5, 0,.5, 0]
e = [ 0, 1,-1, 0]
```

```python
w_sum(a,b)
w_sum(b,c)
w_sum(b,d)
w_sum(c,c)
w_sum(d,d)
w_sum(c,e)
```
